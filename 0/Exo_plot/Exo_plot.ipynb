## 9
Tracer un graphique en boîte (boxplot) à partir des données suivantes :
data = [np.random.normal(0, 1, 100), np.random.normal(2, 1, 100), np.random.normal(-3, 2, 100)]

```python
data = [np.random.normal(0, 1, 100), 
        np.random.normal(2, 1, 100), 
        np.random.normal(-3, 2, 100)]
plt.boxplot(data, 
            notch=False, 
            sym="bo", 
            vert=False, 
            whis=1.5, 
            labels=['set 1',' set 2',' set 3'], 
            meanline=True)
plt.title('Boxplot')
plt.xlabel('Valeurs observées')
plt.ylabel('3 jeux de données différents')
plt.grid()
```

```python
#Boxblot en OO avec couleur ! 
fig, (graph1) = plt.subplots()
graph1.set_title('Boxplot en couleur!')
ma_joli_box = graph1.boxplot(data, 
                             notch=True, 
                             sym="o", 
                             vert=False, 
                             labels=['set 1',' set 2',' set 3'],
                             patch_artist=True)


colors = ['pink', 'lightblue', 'lightgreen']
for patch, color in zip(ma_joli_box['boxes'], colors):
    patch.set_facecolor(color)
    
graph1.xaxis.grid()
graph1.set_xlabel('Valeurs observées')
graph1.set_ylabel('3 jeux de données différents')

plt.show()
```

Tracer un nuage de points à partir de données aléatoires générées par NumPy.


```python
#création des points avec (x,y) aléatoires
data_x = np.random.normal(0,10, 100)
data_y = np.random.normal(0,10,100)
data_z = np.random.normal(0,100,100)
data_quatro= np.random.normal(0,100,100)
plt.scatter(data_x,
            data_y,
            s = data_quatro, #taille des points (arraylike si variable)
            c = data_z, #couleur des points
            marker = "P", #marker std matplotlib: https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
            cmap ='viridis',   #colormap function ou str pour appel
            # norm = 'linear', #une normalisation pour la colormap
            alpha=0.5, #self_explanatory, transparence des points en cas de superposition, float entre 0 et 1
            linewidths=1,  #épaisseur des bords des marqueurs (1.5 par défaut)
            edgecolors='black', #couleurs des bords des marqueurs (face, none, ou une sequence de couleurs)
            plotnonfinite=False #bool, s'il faut tracer les points non définis (NaN ou inf)

            )
```

```python
#Scatter + Histo avec un set aléatoire , en OO

#délimiter la figure
fig = plt.figure(figsize=(6,6))

''' on va définir un "gridspec" de 2x2 (on a trois graph dans un tableau et une case vide)
il est possible de définir le ratio des axes selon les besoins, voir les params de la méthode addgridspec'''
my_grid = fig.add_gridspec(2, 2, #taille de la maille
                           width_ratios = (4, 1),
                           height_ratios = (1, 4),#ratio des axes
                            # left = 0.2,
                            # right = 0.8, 
                            # bottom = 0.2,
                            # top = 0.8, #fractions de la figure d'origine, jouer avec pour voir, left > right, top > bottom, marges avec le bord
                            wspace = 0.2,
                            hspace = 0.2, #marges entre les graphes
                           )

#création des axes

axe = fig.add_subplot(my_grid[1,0]) # premier graph (scatter), dans la seconde ligne premier colonne du gridspec
axe_histx = fig.add_subplot(my_grid[0,0], sharex = axe) #etc 
axe_histy = fig.add_subplot(my_grid[1,1], sharey = axe)

#on trace le scatter
axe.scatter(data_x, data_y)
#on ajoute les histo
axe_histx.hist(data_x, bins = 50, density= True)
axe_histy.hist(data_y, bins = 50, density = True, orientation='horizontal')
# les abcisses des histos sont redondantes, on peut les enlever : 
axe_histx.tick_params(axis='x', labelbottom = False)
axe_histy.tick_params(axis='y', labelleft = False)
```

```python
#si on essaie de rajouter une catégorie à deux jeux de données : 
import matplotlib
taille_h = np.random.randint(150,190,100)
masse_h = np.random.randint(60,120,100)
taille_f = np.random.randint(140,175,100)
masse_f = np.random.randint(40,90,100)
couleurs = np.random.randint(0,2, 100)
fig = plt.figure(figsize=(4,4))

my_grid = fig.add_gridspec(2, 2, 
                           width_ratios = (4, 1),
                           height_ratios = (1, 4),
                            wspace = 0.2,
                            hspace = 0.2,
                           )

axe = fig.add_subplot(my_grid[1,0]) 
axe_histx = fig.add_subplot(my_grid[0,0], sharex = axe) 
axe_histy = fig.add_subplot(my_grid[1,1], sharey = axe)

#pour colorier le scatter et les histo de la même façon
type_couleur= [matplotlib.cm.Accent(0) , matplotlib.cm.Accent(1)]

axe.scatter(x= taille_h, y= masse_h, color = type_couleur[0], label= 'Hommes')
axe.scatter(x= taille_f, y= masse_f, color = type_couleur[1], label = 'Femme')
axe.set_xlabel('Taille (cm)')
axe.set_ylabel('Poids (kg)')

axe_histx.hist([taille_h, taille_f] , color = type_couleur, bins = 50, density= True)

axe_histy.hist([masse_h, masse_f] , color = type_couleur, bins = 50, density = True, orientation='horizontal')

axe_histx.tick_params(axis='x', labelbottom = False)
axe_histy.tick_params(axis='y', labelleft = False)
axe.legend()
```

```python
plt.figure(figsize=(6,6))
fig, ax = plt.subplots(1)
ax.set_aspect(1)
def cercle(r):
    thetha = np.linspace(0, 2*np.pi,100)
    x = r * np.cos(thetha)
    y = r * np.sin(thetha)
    return x, y
    
for i in range(1,5):
    x , y = cercle(i)
    ax.plot(x,y)
```

```python
plt.figure(figsize=(6,6))
fig, ax = plt.subplots(1)
ax.set_aspect(1)
def cercle_modif(r):
    n_points = 500
    thetha = np.linspace(0, 2*np.pi, n_points)
    modif = np.random.randint(1, 3, n_points)
    x =  -modif/100 +  (r+(np.sin(modif/1.1))) * np.cos(thetha)
    y = +modif/10 + (r+(np.sin(modif))) * np.sin(thetha)
    ax.plot(x,y)

for i in range(1):
    cercle_modif(i)
```
