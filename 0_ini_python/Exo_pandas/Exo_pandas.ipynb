```python
#un exemple de heatmap
#index des lignes
idx = ['1', '2', '3', '4']
#noms des colonnes
cols = list('ABCD')

#cr√©ation du dataframe
df = pd.DataFrame([[10, 20, 30, 40], [50, 30, 8, 15],
                   [25, 14, 41, 8], [7, 14, 21, 28]],
                   columns = cols, index = idx)

#correlation : 
corr = df.corr()
corr
```

```python
#pour colorier le dataframe : 
corr.style.background_gradient(cmap='coolwarm')
```
