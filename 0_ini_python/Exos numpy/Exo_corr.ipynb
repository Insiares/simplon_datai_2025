# Veille / Exo sur la correlation 

Avec les mains, la correlation

# Exemple entre deux random set

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.random.randn(50)
y = -3*x + 4 + np.random.randn(50)

#on peut calculer la correlation avec np

corr = np.correlate(x,y)
print(corr)
plt.plot(x,y, '.')
```

```python
#Correlate nous renvoit un facteur liant les deux sets, si on veut un coeff 

coef = np.corrcoef(x,y)
coef
# corrcoef nous renvoit la matrice des coefficients de correlation , ici 
#  [ (x,f(x)), (x, f(y))]
#  [ (y, f(x)), (y, f(y))]

# les coefficients sont proches de 0, la correlation est faible 
# corrcoef renvoit des coefficients de Pearson, qui s'équivaut à une normalisation de la covariance
```

# Exemple avec cos et sin 

```python
theta = np.linspace(0, 2*np.pi, 50)
x = np.sin(theta)
y = np.cos(theta)
corr = np.correlate(x ,y)
print(corr)

coeff = np.corrcoef(x,y)
print(coeff)

# la correlation est quasi-nulle car sin(x) et cos(x) sont de phases inverses
```

# two gaussian distributed data with different mean and variance

```python
x = np.random.default_rng().normal(1, #moyenne
                               2.5, #écart type
                               size=500 )

y = np.random.default_rng().normal(15,
                               4, 
                               size=500 )
plt.hist(x, bins = 50)
plt.hist(y, bins = 50, alpha = 0.5)
corr = np.correlate(x,y)
print(corr)

coeff = np.corrcoef(x,y)
print(coeff)
```

```python
plt.plot(x,y, '.')
```

```python
# si on superpose les deux 
x = np.random.default_rng().normal(1, #moyenne
                               2.5, #écart type
                               size=100 )

y = np.random.default_rng().normal(1,
                               2.5, 
                               size=100 )
plt.hist(x, bins = 50)
plt.hist(y, bins = 50, alpha = 0.5)
corr = np.correlate(x,y)
print(corr)

coeff = np.corrcoef(x,y)
print(coeff)
#ici, correlate nous renvoit une valeur haute, car beaucoup de points en commun
#mais les coeff sont proches de 0
```

```python
plt.plot(x,y, '.')
```
