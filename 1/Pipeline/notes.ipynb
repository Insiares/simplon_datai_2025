```python
import pandas as pd 
import numpy as np
```

- Donnez un nom aux colonnes: ['ID','salle','note']
- Ajoutez une colonne note_dec qui contient les notes de 0 à 17 (E- à A+)
- Convertissez les notes pour qu’elles soient entre 0 et 20 en appliquant une fonctino
- Ajoutez une colonne "result" qui contiendra "Réussite" pour les étudiants ayant une note supérieure ou égale à 10, sinon "Échec".

```python
grades_df = pd.read_csv('student_grades.csv', header = None) #header indique que le csv ne contient pas de titre de colonne
grades_df.columns = ['ID','salle','note']
grades_df
```

```python
note_us = ['E-', 'E', 'E+', 'D-', 'D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+']
note_norm = [y for y in range(15)]
conversion = dict(zip(note_us, note_norm))
```

```python
grades_df.insert(loc=3, column='note_dec', value=grades_df['note'])
grades_df['note_dec'] = grades_df['note'].replace(conversion)
grades_df
```

```python
def note_20(note):
    ratio = note/14
    new_note = round(ratio * 20,2)
    return new_note

grades_df['note_dec'] = grades_df['note_dec'].apply(note_20)
grades_df
```

```python
grades_df.insert(loc = 4, column= 'Résultat', value = 'Echec')
grades_df
```

```python
grades_df['Résultat'] = grades_df['Résultat'].mask(grades_df['note_dec']>=10, 'Réussite')
#grades_df['Résultat'] = grades_df['Résultat'].mask(grades_df['note_dec']<10, 'Echec')
grades_df
```
