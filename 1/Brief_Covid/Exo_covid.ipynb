Quelle est la probabilit√© pour un Fran√ßais (vs) une Fran√ßaise de mourir du Covid au d√©but de la p√©riode disponible ? √† la fin de la p√©riode disponible ?

```python
#Calcul proba
def proba_dc(idx1, data, population):
    total= population
    dc = data[idx1:idx1+1]['dc']
    Proba = 100*(dc/total)
    return Proba
```

```python
proba = []
for i in range(df.shape[0]):
    proba.append(proba_dc(i,df, 68_000_000))

plt.plot(df['Unnamed: 0'], proba, label = 'Hosp, homme')
```

```python
fig, axe1 = plt.subplots(figsize=(15,6))

fraction_1 = 100*(df['dc.1'][::40]) / 32_920_000
axe1.plot(df['Unnamed: 0'][::40], (fraction_1), label = 'Homme', color = 'blue')
axe2 = axe1.twinx()
fraction_2 = 100*(df['dc.2'][::40]) / 35_122_000
axe2.plot(df['Unnamed: 0'][::40], (fraction_2), label = 'Femme', color = 'red')
#fig.tight_layout()
axe1.set_xlabel('Date')
axe1.set_ylabel("Probabilit√©(%) de d√©c√®s, Homme (log) ")
axe2.set_ylabel('Probabilit√©(%) de d√©c√®s, Femme (log)')
axe1.tick_params(axis='x', labelrotation = 90, labelsize=10)
axe1.set_yscale('log')
axe2.set_yscale('log')
axe1.set_ylim(0, 0.3)
axe2.set_ylim(0, 0.3)
plt.rc('axes', titlesize=16)        # Controls Axes Title
plt.rc('axes', labelsize=16)        # Controls Axes Labels
plt.rc('xtick', labelsize=16)       # Controls x Tick Labels
plt.rc('ytick', labelsize=16)       # Controls y Tick Labels
plt.rc('legend', fontsize=16)       # Controls Legend Font
plt.rc('figure', titlesize=16)      # Controls Figure Title
plt.title('Probablit√©(%) de deces en France en fonction du temps')
plt.vlines(df['Unnamed: 0'][284],0, 0.25, colors='black', label = '1er vaccination')
plt.vlines(df['Unnamed: 0'][534],0, 0.25, colors='grey', label = '1er rappel')
fig.legend(bbox_to_anchor=(0.9, 0.5))

```
